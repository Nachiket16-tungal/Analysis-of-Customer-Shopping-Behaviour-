```python
df["Review Rating"]= df.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()                 #to convert column name to lower case
df.columns=df.columns.str.replace(' ','_')        # to remove space and replacing them with "_"
df=df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})   #to remove usd from purchase amount columnn this code is written
```

```python
df.columns
```

```python
#create column age_group
labels=["young_adult","adult","middle_aged","senior"]
df["age_group"]=pd.qcut(df["age"], q=4,labels=labels) 
```

```python
df[["age","age_group"]].head(10)
```

```python
# create column purchase_frequency_days
frequency_mapping={
    "Fortnightly":14,
    "Weekly":7,
    "Quarterly":90,
    "Bi-Weekly":14,
    "Annually":365,
    "Every 3 Months":90
}
df["purchase_frequency_days"]=df["frequency_of_purchases"].map(frequency_mapping)
```

```python
df[["purchase_frequency_days","frequency_of_purchases"]].head(10)
```
